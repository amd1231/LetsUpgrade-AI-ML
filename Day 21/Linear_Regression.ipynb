
# Linear Regression between Price and SQFT

```python
price = dataset.iloc[ : , :1 ]
sqft = dataset.iloc[ : , 1:2 ]
```

```python
plt.scatter(sqft,price)
plt.title("SQFT vs PRICE\n")
plt.xlabel("Sqft_living")
plt.ylabel("Price")
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_test,X_train,Y_test,Y_train = train_test_split(sqft,price,test_size=1/3,random_state=3)
```

```python
 from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
model.fit(X_train,Y_train)
```

```python
model.coef_
```

```python
model.intercept_
```

```python
plt.scatter(X_train,Y_train,color = 'red')
plt.plot(X_train,model.predict(X_train),color = 'green')
plt.title("Price Variation\n")
plt.xlabel("Sqft_living")
plt.ylabel("Price")
```

```python
Y_pred = model.predict(X_test)
```

```python
plt.scatter(X_test,Y_test,color = 'green')
plt.plot(X_test,model.predict(X_test),color = 'blue')
plt.title("Price Variation\n")
plt.xlabel("Sqft_living")
plt.ylabel("Price")
```

```python
from sklearn.metrics import r2_score,mean_squared_error
import numpy as np
```

```python
RMSE = np.sqrt(mean_squared_error(Y_test, Y_pred))
r_square = r2_score(Y_test, Y_pred)
```

```python
print("The RMSE value is : " ,RMSE)
print("RSuare value is :  ",r_square)
```
