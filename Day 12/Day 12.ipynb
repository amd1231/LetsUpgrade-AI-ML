```python
import pandas as pd
dataset = pd.read_csv("general_data.csv")

from sklearn import preprocessing
encoder = preprocessing.LabelBinarizer()
dataset["Attrition"] = encoder.fit_transform(dataset["Attrition"]) 
```

```python
dataset.head()
```

```python
dataset.columns
```

```python
dataset1=dataset.dropna()
```

```python
dataset1
```

```python
from scipy.stats import mannwhitneyu
dataset1.head()
```

```python
Attrition_YES = dataset1.groupby("Attrition")
Attrition_YES = Attrition_YES.get_group(1)
Attrition_YES.head()
```

```python

Attrition_No = dataset1.groupby("Attrition")
Attrition_No = Attrition_No.get_group(0)
Attrition_No.head()
```

```python
dataset1.columns
```

# Attrition VS Distance from Home

```python
stats,p =mannwhitneyu(Attrition_YES.DistanceFromHome,Attrition_No.DistanceFromHome)
```

```python
print(stats,p)
null_hypothesis = "There is no significant relation between attrition and distance from home"
alternate_hypothesis = "There is a significant relation between attrition and distance from home"
if p < 0.05:
    print(alternate_hypothesis)
else:
    print(null_hypothesis)
```

```python
stats,p = mannwhitneyu(Attrition_YES.MonthlyIncome,Attrition_No.MonthlyIncome)
print(stats,p)
null_hypothesis = "There is no significant relation between attrition and Monthly Income"
alternate_hypothesis = "There is a significant relation between attrition and Monthly Income"
if p < 0.05:
    print(alternate_hypothesis)
else:
    print(null_hypothesis)
```


# CHI2 TEST

```python
from scipy.stats import chi2_contingency
```

```python
chitable = pd.crosstab(dataset1.Attrition,dataset1.Gender)
```

```python
print(chitable)
```

```python
stats,p,dof,expected = chi2_contingency(chitable)
print(stats,p)
if p < 0.05:
    print("Null Hypothesis rejected")
else:
    print("Alternate Hypothesis accepted")
```

```python
chitable1 = pd.crosstab(dataset1.Attrition,dataset1.BusinessTravel)
print(chitable1)
```

```python
stats,p,dof,expected = chi2_contingency(chitable1)
print(stats,p)
if p < 0.05:
    print("Null Hypothesis rejected")
else:
    print("Alternate Hypothesis accepted")
```
